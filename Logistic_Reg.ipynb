```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('bank_full.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,columns=['job','marital','education','contact','poutcome','month'])
```

```python
dataset.info()
```

```python
x_d = pd.concat([dataset.iloc[:,0:10],dataset.iloc[:,11:]],axis=1)
y = dataset.iloc[:, 10].values
```

```python
x_d['default'] = np.where(x_d['default'].str.contains("yes"), 1, 0)
x_d['housing'] = np.where(x_d['housing'].str.contains("yes"), 1, 0)
x_d['loan'] = np.where(x_d['loan'].str.contains("yes"), 1, 0)
x_d
```

```python
x=x_d.values
```

```python
print(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

### Comparing actual result of y_train and y_pred which has predicted data from y_train dataset

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
